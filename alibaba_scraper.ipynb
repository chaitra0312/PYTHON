```python
import asyncio
from playwright.async_api import async_playwright
async def run():
    async with async_playwright() as p:
        browser = await p.chromium.launch(headless=True)
        page = await browser.new_page()
        await page.goto("https://sourcing.alibaba.com/rfq/rfq_search_list.htm", timeout=60000)
        await page.wait_for_timeout(5000)
        all_divs = await page.query_selector_all("div")
        print("Looking for RFQ-related classes...\n")
        found = False
        for div in all_divs:
            class_name = await div.get_attribute("class")
            if class_name and "rfq" in class_name.lower():
                found = True
                text = await div.inner_text()
                print(f"[{class_name}]\n{text[:300]}...\n")

        if not found:
            print("No RFQ-related elements found.")

        await browser.close()

await run()
```

```python
import asyncio
from playwright.async_api import async_playwright
async def run():
    async with async_playwright() as p:
        browser = await p.chromium.launch(headless=True)
        page = await browser.new_page()
        await page.goto("https://sourcing.alibaba.com/rfq/rfq_search_list.htm", timeout=60000)
        await page.wait_for_timeout(5000)
        rfq_cards = await page.query_selector_all(".brh-rfq-item")
        print(f"Found {len(rfq_cards)} RFQ cards.\n")
        for i, card in enumerate(rfq_cards[:5]):
            title = await card.query_selector(".brh-rfq-item__main-info")
            desc = await card.query_selector(".brh-rfq-item__detail")
            qty = await card.query_selector(".brh-rfq-item__quantity")
            country = await card.query_selector(".brh-rfq-item__country")
            date = await card.query_selector(".brh-rfq-item__publishtime")
            buyer = await card.query_selector(".brh-rfq-item__other-info")
            print(f"--- RFQ {i+1} ---")
            print("Title:", await title.inner_text() if title else "N/A")
            print("Description:", await desc.inner_text() if desc else "N/A")
            print("Quantity:", await qty.inner_text() if qty else "N/A")
            print("Country:", await country.inner_text() if country else "N/A")
            print("Date Posted:", await date.inner_text() if date else "N/A")
            print("Buyer Info:", await buyer.inner_text() if buyer else "N/A")
            print()
        await browser.close()
await run()
```

```python
import asyncio
import csv
from playwright.async_api import async_playwright
async def run():
    data = []
    async with async_playwright() as p:
        browser = await p.chromium.launch(headless=True)
        page = await browser.new_page()
        await page.goto("https://sourcing.alibaba.com/rfq/rfq_search_list.htm", timeout=60000)
        await page.wait_for_timeout(5000)
        rfq_cards = await page.query_selector_all(".brh-rfq-item")
        for card in rfq_cards[:10]:  # Limit to 10 for now
            title = await card.query_selector(".brh-rfq-item__main-info")
            desc = await card.query_selector(".brh-rfq-item__detail")
            qty = await card.query_selector(".brh-rfq-item__quantity")
            country = await card.query_selector(".brh-rfq-item__country")
            date = await card.query_selector(".brh-rfq-item__publishtime")
            buyer = await card.query_selector(".brh-rfq-item__other-info")
            data.append({
                "Title": await title.inner_text() if title else "N/A",
                "Description": await desc.inner_text() if desc else "N/A",
                "Quantity": await qty.inner_text() if qty else "N/A",
                "Country": await country.inner_text() if country else "N/A",
                "Date Posted": await date.inner_text() if date else "N/A",
                "Buyer Info": await buyer.inner_text() if buyer else "N/A"
            })
        await browser.close()
    csv_file = "alibaba_rfq_data.csv"
    with open(csv_file, mode="w", newline='', encoding="utf-8") as f:
        writer = csv.DictWriter(f, fieldnames=data[0].keys())
        writer.writeheader()
        writer.writerows(data)
    print(f"✅ Data saved to '{csv_file}' with {len(data)} rows.")
await run()
```
